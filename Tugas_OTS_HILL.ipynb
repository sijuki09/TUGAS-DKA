### PENERAPAN DALAM BENTUK CODE PROGRAM

```python
import random

# Representasi jarak antar kota (matriks simetris)
distance_matrix = [
    # A   B   C   D   E
    [ 0, 10, 15, 20, 25],  # A
    [10,  0, 35, 25, 15],  # B
    [15, 35,  0, 30, 20],  # C
    [20, 25, 30,  0, 10],  # D
    [25, 15, 20, 10,  0],  # E
]

def total_distance(tour):
    dist = 0
    for i in range(len(tour)):
        from_city = tour[i]
        to_city = tour[(i + 1) % len(tour)]  # kembali ke awal
        dist += distance_matrix[from_city][to_city]
    return dist

def get_neighbors(tour):
    neighbors = []
    for i in range(1, len(tour) - 1):  # jangan tukar posisi 0 (A) agar tetap jadi titik awal
        for j in range(i + 1, len(tour)):
            neighbor = tour[:]
            neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
            neighbors.append(neighbor)
    return neighbors

# Nama kota untuk ditampilkan
city_names = ['A', 'B', 'C', 'D', 'E']

# Inisialisasi solusi awal
current_solution = [0, 1, 2, 3, 4]  # Mulai dari A
random.shuffle(current_solution[1:])  # Acak selain posisi pertama
current_distance = total_distance(current_solution)

print("Solusi awal:", [city_names[i] for i in current_solution], "| Total jarak:", current_distance)

# Hill Climbing
while True:
    neighbors = get_neighbors(current_solution)
    best_neighbor = min(neighbors, key=total_distance)
    best_distance = total_distance(best_neighbor)

    if best_distance >= current_distance:
        break

    current_solution = best_neighbor
    current_distance = best_distance
    print("Ditemukan solusi lebih baik:", [city_names[i] for i in current_solution], "| Jarak:", current_distance)

# Cetak hasil akhir
print("\n=== HASIL AKHIR ===")
print("Rute terbaik:", ' → '.join([city_names[i] for i in current_solution + [current_solution[0]]]))
print("Total jarak terbaik:", current_distance)
```
